```python


pdf_list = []
import PyPDF2
pdfFileObject = open('Migrants_Pour_System.pdf', 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObject)
count = pdfReader.numPages
for i in range(count):
    page = pdfReader.getPage(i)
    page.extractText().split()
    pdf_list.append(page.extractText()) #How read these better?
```

```python
String_pdf= str(pdf_list)
#String_pdf = String_pdf[:]
print(String_pdf)
```

Step 2:

Create Stopwords and punctuations list in order to remove it from the original text since they don't hold much meaning

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
nltk.download('punkt')

#The word_tokenize() function will break our text phrases into #individual words
tokens = word_tokenize(String_pdf)
#we'll create a new list which contains punctuation we wish to clean
punctuations = ['(',')',';',':','[',']',',']
#We initialize the stopwords variable which is a list of words like #"The", "I", "and", etc. that don't hold much value as keywords
stop_words = stopwords.words('english')
#We create a list comprehension which only returns a list of words #that are NOT IN stop_words and NOT IN punctuations.
keywords_not_cleaned = [word for word in tokens if not word in stop_words and not word in punctuations]
```

```python
keywords_not_wanted=keywords_not_cleaned[1:169] #How Do I remove this from my original list?
keywords_not_wanted
```

```python
def Removing_not_wanted_keywords(list1):
    del list1[0:169]
    return (list1) #If you run it a lot of times it keeps deleting the first 169 objects.
```

```python
keywords=Removing_not_wanted_keywords(keywords_not_cleaned)
```

```python
keywords
```

```python

sentences_newspaper = []
list_sentiment_token = []
blob = TextBlob(str(keywords))
blob.tags         

blob.noun_phrases  

#print(blob.sentences)

for sentence in blob.sentences:
    
    sentence.sentiment.polarity   #These are the polarity (sentiment that-1 to 1) of each sentence in the PDF file
    list_sentiment_token.append(sentence.sentiment.polarity)   
    sentences_newspaper.append(sentence)
    
# I expect that if the library returns exactly 0.0, 
#then your sentence didn't contain any words that had a polarity in the NLTK training set. 
#I suspect the researchers didn't include them because 1) they were too rare in the training data or 2) 
#they were known to be meaningless (such as "the", "a", "and", etc.).
```

```python
float_list_sentiment=[float(i) for i in list_sentiment_token] 

ln = len(sentences_newspaper)
print(ln)
lt=len(list_sentiment_token)
print(lt)
dictionary_sentences_score = dict(zip( float_list_sentiment , sentences_newspaper))
ld = len(dictionary_sentences_score)
print(ld)# we can see that they are not the same len
```
